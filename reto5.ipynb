```python
import pandas as pd

rutaFileCsv = 'https://github.com/luisguillermomolero/MisionTIC2022_2/blob/master/Modulo1_Python_MisionTIC2022_Main/Semana_5/Reto/movies.csv?raw=true'

def listaPeliculas(path: str) -> str:
  
  if rutaFileCsv.split('.')[-1].find('csv') != 0:
    return 'Extensión inválida.'
  
  try:
    dfFile = pd.read_csv(path)
    dfPeliculas = dfFile[['Country', 'Language', 'Gross Earnings']]
    ganancia = dfPeliculas.pivot_table(index = ['Country', 'Language'])
    return ganancia.head(10)

  except:
    return 'Error al leer el archivo de datos.'


print(listaPeliculas(rutaFileCsv))
```
