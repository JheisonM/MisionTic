```python
def ordenes(rutinaContable):
  print('------------------------ Inicio Registro diario ---------------------------------')
 
  for factura in rutinaContable:
    
    cantidad = 0.0
    
    for index in range(1, len(factura)):
      producto = factura[index]
      cantidad += producto[1] * producto[2]
      
    if cantidad < 600000: cantidad += 25000
    
    print(f'La factura {factura[0]} tiene un total en pesos de {cantidad:,.2f}')
 
  print('-------------------------- Fin Registro diario ----------------------------------')

rutinaContable = [ [1201, ("5464", 4, 25842.99), ("7854",18,23254.99), ("8521", 9, 48951.95)],
                   [1202, ("8756", 3, 115362.58),("1112",18,2354.99)],
                   [1203, ("2547", 1, 125698.20), ("2635", 2, 135645.20), ("1254", 1, 13645.20),("9965", 5, 1645.20)],
                   [1204, ("9635", 7, 11.99), ("7733",11,18.99), ("88112", 5, 390.95)]
]
ordenes(rutinaContable)
```
