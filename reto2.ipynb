```python
def cliente(informacion:dict) -> dict :

  retorno = {'nombre': informacion['nombre'], 
             'edad': informacion['edad'],  
             'atraccion': 'N/A', 
             'apto': False, 
             'primer_ingreso': informacion['primer_ingreso'], 
             'total_boleta': 'N/A'}

  descuento = 0.0

  if informacion['edad'] >= 7 :
    retorno['apto'] = True
    if informacion['edad'] < 15 :
      retorno['atraccion'] = 'Sillas voladoras'
      retorno['total_boleta'] = 10000
      descuento = 0.05
    elif informacion['edad'] <= 18 :
      retorno['atraccion'] = 'Carros chocones'
      retorno['total_boleta'] = 5000
      descuento = 0.07
    else : 
      retorno['atraccion'] = 'X-Treme'
      retorno['total_boleta'] = 20000
      descuento = 0.05

    if informacion['primer_ingreso'] == True :
      retorno['total_boleta'] -= retorno['total_boleta'] * descuento
    
  return retorno


entrada = {"id_cliente": 1, "nombre": 'Johana Fernandez', "edad": 3, "primer_ingreso": True}
print(cliente(entrada))
```
